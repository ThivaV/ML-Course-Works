```python
# new column from list (if We using list it has to be same length)
# if it is Series, length doesn't matter because not assigned cells assigned NaN

fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 4.7, 7.6, 8.7, 3.0, 4.5]
car_sales["Fuel per 10KM"] = fuel_economy
car_sales
```

```python
car_sales["Total fuel used (L)"] = car_sales["Odometer (KM)"] / 100 * car_sales["Fuel per 10KM"]
```

```python
car_sales
```

```python
# Create a column from a single value
car_sales["Number of wheels"] = 4
car_sales
```

```python
car_sales["passed road saftey"] = True
car_sales
```
